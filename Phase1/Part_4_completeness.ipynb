# **Orders Dataset**

```
df_orders = pd.read_csv('orders.csv')
df_orders
```

\\
## Method 1: 

$$ completeness = \frac {records\ with\ no\ missing\ attribute}{Total\ records\ in\ a\ dataset}$$

```
noNan_df = df_orders.dropna(axis=0)
noNan_df
```

```
completeness_percent = len(noNan_df) / len(df_orders) * 100
print(f"completeness with method 1 : {completeness_percent} %" )
```

## Method 2

$$ completeness = \frac {A: number\ of\ data\ required\ for\ the\ particular\ context\ in\ the\ data file}{B: number\ of\ data\ in\ the\ specified\ particular\ context\ of\ intended\ use}$$

We have selected 2 more important columns and stored them in new dataframe.

```
new_df_orders = df_orders[['Amount_Gross_Order', 'Quantity_item']]
new_df_orders
```

delete rows which have at least 1 Nan.

```
noNan_new_df = new_df_orders.dropna(axis=0)
noNan_new_df
```

```
completeness_percent = len(noNan_new_df) / len(df_orders) * 100
print(f"completeness with method 2 :{completeness_percent} %" )
```

## Method 3

$$ completeness = \frac {A: attribute\ fields\ containing\ values}{B: records\ \times attributes}$$

```
number_of_notnulls = df_orders.notnull().sum().sum()
print("Total number of null cells: ", number_of_notnulls)
```

```
completeness_percent = number_of_notnulls / (df_orders.shape[0] * df_orders.shape[1]) * 100
print(f"completeness with method 3 :{completeness_percent} %" )
```
